```python
import os
import random
import shutil

# Set paths
base_dir = "dataset"
img_dir = os.path.join(base_dir, "images")
label_dir = os.path.join(base_dir, "labels")

# New dirs
train_img_dir = os.path.join(base_dir, "images/train")
val_img_dir = os.path.join(base_dir, "images/val")
train_lbl_dir = os.path.join(base_dir, "labels/train")
val_lbl_dir = os.path.join(base_dir, "labels/val")

# Create dirs
os.makedirs(train_img_dir, exist_ok=True)
os.makedirs(val_img_dir, exist_ok=True)
os.makedirs(train_lbl_dir, exist_ok=True)
os.makedirs(val_lbl_dir, exist_ok=True)

# Split
image_files = [f for f in os.listdir(img_dir) if f.endswith(('.jpg', '.jpeg', '.png', '.webp'))]
random.shuffle(image_files)

split_ratio = 0.8
split_idx = int(len(image_files) * split_ratio)
train_files = image_files[:split_idx]
val_files = image_files[split_idx:]

# Move files
for img_file in train_files:
    shutil.move(os.path.join(img_dir, img_file), os.path.join(train_img_dir, img_file))
    lbl_file = img_file.rsplit(".", 1)[0] + ".txt"
    if os.path.exists(os.path.join(label_dir, lbl_file)):
        shutil.move(os.path.join(label_dir, lbl_file), os.path.join(train_lbl_dir, lbl_file))

for img_file in val_files:
    shutil.move(os.path.join(img_dir, img_file), os.path.join(val_img_dir, img_file))
    lbl_file = img_file.rsplit(".", 1)[0] + ".txt"
    if os.path.exists(os.path.join(label_dir, lbl_file)):
        shutil.move(os.path.join(label_dir, lbl_file), os.path.join(val_lbl_dir, lbl_file))


```
